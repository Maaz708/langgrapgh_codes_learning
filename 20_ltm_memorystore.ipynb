```python
from langgraph.store.memory import InMemoryStore
from langchain_huggingface import HuggingFaceEmbeddings
from dotenv import load_dotenv
load_dotenv()
```

```python
store = InMemoryStore()
```

```python
namespace=("user","u1")
```

Creating memories

```python
store.put(namespace,"1",{"data":"User likes pizza"})
store.put(namespace,"2",{"data":"User prefers dark mode"})
```

```python
namespace2 = ("user","u2")

store.put(namespace2,"1",{"data":"User likes burger"})
store.put(namespace2,"2",{"data":"User prefers bike"})
```

Retrieving Memories

```python
store.get(namespace,"1")
```

```python
print(store.get(namespace2,"1"))
print(store.get(namespace2,"2"))
```

Retrieving all memories

```python
items = store.search(namespace)
for item in items:
    print(item.value)
```

Semantice search

```python
embeddings = HuggingFaceEmbeddings(model="BAAI/bge-large-en-v1.5")
```

```python
store = InMemoryStore(index={"embed":embeddings,'dims':384})
```

```python
namespace=("users","u1")
```

```python
store.put(namespace, "1", {"data": "User prefers action movies"})

store.put(namespace, "2", {"data": "User prefers pizza on weekends"})
store.put(namespace, "3", {"data": "User drinks coffee every morning"})
store.put(namespace, "4", {"data": "User enjoys playing cricket"})
store.put(namespace, "5", {"data": "User is learning Python and LangGraph"})
store.put(namespace, "6", {"data": "User likes watching sci-fi movies"})
```

```python
items = store.search(namespace,query="What is user currently learning", limit=1)
for item in items:
    print(item.value)
```

```python
items = store.search(namespace,query="What is user's preferances", limit=3)
for item in items:
    print(item.value)
```
